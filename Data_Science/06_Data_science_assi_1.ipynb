# Turn categorical variables into quantitative variables 

```python
# Assuming 'df' is your DataFrame
df['Purchased Bike'] = df['Purchased Bike'].replace(['Yes', 'No'], [1, 0])

# Convert the column to numeric
df['Purchased Bike'] = pd.to_numeric(df['Purchased Bike'], errors='coerce')

# Display the DataFrame
df
```

# Conversion of int into float

```python
df1=df['Purchased Bike'].astype(float)
df['Purchased Bike']=df1
df1
```
