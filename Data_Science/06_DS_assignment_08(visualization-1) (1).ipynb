```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= sns.load_dataset('titanic')
```

```python
df=df.drop('deck',axis=1)
```

```python
df['age']=df['age'].fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df['age'])
plt.show()
```

# catplot - Gender , survived

```python
sns.catplot(x='sex',y='survived',data=df)
plt.show()
```

# violin :- genders , age (split=true)

```python
sns.violinplot(x='sex',y='age',split=True,data=df)
plt.show()
```

# barplot - (fare , survived)

```python
sns.barplot(x='fare',y='survived',data=df)
plt.show()
```

# distplot(fare)

```python
    sns.displot(x='fare',data=df)
    plt.show()
```

# jointplot(age,fare,kind=?)

```python
sns.jointplot(x='age',y='fare',kind='reg',data=df)
plt.show()
```

```python
sns.jointplot(x='age',y='fare',kind='kde',data=df)
plt.show()
```

```python
sns.jointplot(x='age',y='fare',kind='scatter',data=df)
plt.show()
```

```python
sns.jointplot(x='age',y='fare',kind='resid',data=df)
plt.show()
```

```python
sns.jointplot(x='age',y='fare',kind='hex',data=df)
plt.show()
```

# Groupby - pclass , survived(heatmap :count (not correlation ))
   

```python
grouped_df=df.groupby(['pclass','survived']).size().reset_index(name='count')
pivot_df = grouped_df.pivot_table(index='pclass',columns='survived',values='count')

sns.heatmap(pivot_df,annot=True,fmt="d",cmap="YlGnBu")
plt.title('Survivors of pclass')
plt.xlabel('Surviors')
plt.ylabel('Pclass')
plt.show()
```

### Write a code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram


```python
plt.hist(df['fare'],bins=20,color='skyblue',edgecolor='black')
plt.xlabel('fare')
plt.ylabel('Frequency')
plt.title('distribution of ticket prices of passengers')
plt.grid(True)
plt.show()
```
