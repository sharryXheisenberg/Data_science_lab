# Calculating outliers through IQR method

```python
import seaborn as sns
```

```python
sns.boxplot(df1['dis'])
```

```python
np.where(df1['dis']>10)
```

```python
df1.shape
```

```python
Q1 = np.percentile(df1['dis'],25,interpolation ='midpoint')
Q3 = np.percentile(df1['dis'],75,interpolation ='midpoint')
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
high = np.where(df1['dis']>=Q3 + 1.5*IQR)
low = np.where(df1['dis']<=Q1 - 1.5*IQR)
```

```python
print(high)
print(low)
```

```python
df1.drop(high[0],inplace = True)
df1.drop(low[0],inplace = True)
```

# Standardization

```python
dict1 = {"Roll no":[1,2,3,4,5],
         "Name":['A' , 'B' , 'C' , 'D', 'E'],
         "Maths marks":[60,70,80,90,95],
         "English marks":[70,80,90,78,60],
         "Science marks":[80,90,60,60,70],
         "History marks":[90,50,60,70,80],
         "Geography marks":[50, 60,70,80,90]
        }

df = pd.DataFrame(dict1)
```

```python
df
```

```python
df.drop('Name', axis=1, inplace=True)


df
```

```python
df
```

```python
## scaling
from sklearn.preprocessing import MinMaxScaler


columns_to_scale = ['Maths marks', 'Science marks', 'History marks', 'Geography marks']
scaled_df1 = df.copy()
scaled_df1[columns_to_scale] = MinMaxScaler().fit_transform(df[columns_to_scale])


print(scaled_df1)
```

```python
# Plot and visualize the results
import matplotlib.pyplot as plt

df.plot(kind='kde')
plt.title('Before')

scaled_df1.plot(kind='kde')
plt.title('After')
```

```python
## Standardization

from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()

df_scaled = std_scaler.fit_transform(df.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=['Maths marks', 'English marks', 'Science marks', 'History marks', 'Geography marks'])

print("Scaled Dataset Using StandardScaler")
df_scaled.head()
```
